### Testing out `sanjaya` with a video example

```python
import os

from sanjaya import Agent
from sanjaya.tools.video import VideoToolkit

from pydantic_ai.providers.openrouter import OpenRouterProvider
from sanjaya import Agent

agent = Agent(
    model="openai/gpt-5.3-codex",
    sub_model="openai/gpt-4.1-mini",
    provider=OpenRouterProvider(api_key=os.getenv("OPENROUTER_API_KEY")),
)
```

```python
vt = VideoToolkit(max_frames_per_clip=3)
agent.use(vt)

answer = agent.ask(
    'What topics are discussed in this video? Summarize the key points.',
    video='../data/youtube/qdfwmYTO0Aw.mp4',
    subtitle='../data/youtube/meta/qdfwmYTO0Aw_en.json'
)

print(f'\n=== ANSWER ===\n{answer.text}')
print(f'\n=== STATS ===')
print(f'Iterations: {answer.iterations}')
print(f'Cost: \${answer.cost_usd:.6f}')
print(f'Tokens: {answer.input_tokens} in / {answer.output_tokens} out')
print(f'Wall time: {answer.wall_time_s}s')
print(f'Evidence clips: {len(answer.evidence)}')
```
